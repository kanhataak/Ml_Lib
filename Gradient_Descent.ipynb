```python
import numpy as np
from tabulate import tabulate
```

```python
def gradient_descent(x,y):
    m = b = 0
    learning_rate =0.001
    iteration = 100
    n=len(x)
    lst=[("Iteration",'Coefficent','Intercept','Cost')]
    for i in range(iteration):
        
        y_prediction = m*x+b
        cf = (1/n)*sum(y-y_prediction)**2
        md=(-2/n)*sum(x*(y-y_prediction))
        bd=(-2/n)*sum(y-y_prediction)
        m= m- learning_rate*md
        b= b- learning_rate*bd
        
        lst.append((i,m,b,cf))
#         print(tabulate(f"i {i}:,coefficent {m}:, intercept {b}:, cost {cf}:))
    print(tabulate(lst,tablefmt="fancy_grid"))
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
