```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\MR.INDIA\Desktop\Machine Learning\DataSets\townprice.csv")
df
```

```python
dumm=pd.get_dummies(df.town)
dumm
```

```python
mer=pd.concat([df,dumm],axis=1)
mer
```

```python
final=mer.drop("town",axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(final[['area','monroe township','robinsville','west windsor']],final.price)
```

```python
reg.predict([[3000,0,1,0]])
```

```python
reg.predict([[4200,1,0,0]])
```

```python
reg.score(final[['area','monroe township','robinsville','west windsor']],final.price)
```

```python
import seaborn as sn
sn.pairplot(df1)
```

```python
x=df1.drop("salary",axis=1)
y=df1.salary
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
reg1=LogisticRegression()
reg1.fit(x_train,y_train)
```

```python
reg1.predict(x_test)
```

```python
yp=reg1.predict(x_test)
```

```python
reg1.predict([[0.38,0.53,2,157,3,0,1,0,7]])
```

```python
df1
```

```python
from sklearn.metrics import confusion_matrix
kt= confusion_matrix(y_test,yp)
```

```python
kt
```

```python
plt.figure(figsize=(9,6))
sn.heatmap(kt,annot = True)
plt.xlabel("PREDICTED")
plt.ylabel("TRUTH_OR_ACTUAL_VALUE")
```
